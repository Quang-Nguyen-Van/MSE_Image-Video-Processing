```python
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt
import cv2
import numpy as np
from skimage.util import random_noise

data_path = '/content/data/'
```

```python
# Load image
Img = cv2.imread(data_path + 'quadnight.jfif')  # Load image
Img = cv2.cvtColor(Img, cv2.COLOR_BGR2GRAY)    # Convert image to gray scale
Img = cv2.normalize(Img.astype('float'), None, 0.0, 1.0, cv2.NORM_MINMAX) # convert image from integer to float

# Repeat over number of sample images
nArray =[1, 2, 8, 16, 32, 64]

fig = plt.figure(figsize=(20, 20))

for i in range(0, len(nArray)):
  N = nArray[i]
  avgImg = np.zeros(Img.shape, dtype=float) # initialize an empty image

  for j in range(0, N):
    noisyImg = random_noise(Img, mode='gaussian', seed=None, clip=True)
    avgImg = cv2.add(avgImg, noisyImg)

  fig.add_subplot(2, 3, i+1)
  plt.imshow(avgImg, cmap='gray')
  plt.axis('off')
  plt.title(str(N) + ' image(s)')



```
