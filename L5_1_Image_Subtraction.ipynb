```python
from google.colab.patches import cv2_imshow
import cv2
import numpy as np

data_path = '/content/data/'
```

```python
# Load images
Img1 = cv2.imread(data_path + 'live.jpg')
Img2 = cv2.imread(data_path + 'mask.jpg')

# Convert to gray scale
gray_Img1 = cv2.cvtColor(Img1, cv2.COLOR_BGR2GRAY)
gray_Img2 = cv2.cvtColor(Img2, cv2.COLOR_BGR2GRAY)

# Calculate the difference between two images
imgDiff = cv2.absdiff(gray_Img1, gray_Img2)
_, threshold = cv2.threshold(imgDiff, 5, 255, cv2.THRESH_BINARY)

gamma = 2.0
diffContrasted = np.power(imgDiff, gamma)

# Show images
horizontal_concate = np.concatenate((gray_Img1, gray_Img2, diffContrasted), axis=1)
cv2_imshow(horizontal_concate)

# Save image
cv2.imwrite('diffContrasted.jpg', diffContrasted)
```
