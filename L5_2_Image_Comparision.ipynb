```python
from google.colab.patches import cv2_imshow
import cv2
import numpy as np

data_path = '/content/data/'
```

```python
# Load images
maskImg1 = cv2.imread(data_path + 'mask1.png')
maskImg2 = cv2.imread(data_path + 'mask2.png')

# Convert the images to gray scale
gray_maskImg1 = cv2.cvtColor(maskImg1, cv2.COLOR_BGR2GRAY)
gray_maskImg2 = cv2.cvtColor(maskImg2, cv2.COLOR_BGR2GRAY)

# Calculate the difference between two images
diffImg = np.abs(gray_maskImg1 - gray_maskImg2)

# Enhence the contrast
gamma = 2.0
diffImg_Contrasted = np.power(diffImg, gamma)

# Concatenate the images into one image to show
split_img = np.full([diffImg.shape[0], 5], 255)
horizontal_concatenated_image = np.concatenate((diffImg, split_img, diffImg_Contrasted, split_img, gray_maskImg1, split_img, gray_maskImg2), axis=1)

# Show images
cv2_imshow(horizontal_concatenated_image)
cv2.imwrite('Mask_Comparison_Diff.png', diffImg_Contrasted)
```
